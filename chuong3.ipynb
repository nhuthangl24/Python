```python
filename = 'alice.txt'  # T√™n file c·∫ßn ƒë·ªçc

try:
    # M·ªü file v√† ƒë·ªçc n·ªôi dung
    with open(filename, 'r', encoding='utf-8') as f_obj:
        contents = f_obj.read()
except FileNotFoundError:
    # X·ª≠ l√Ω khi file kh√¥ng t·ªìn t·∫°i
    msg = "‚ùå File '" + filename + "' kh√¥ng t·ªìn t·∫°i."
    print(msg)
else:
    # N·∫øu file h·ª£p l·ªá th√¨ ƒë·∫øm s·ªë t·ª´
    words = contents.split()       # T√°ch c√°c t·ª´ trong n·ªôi dung
    num_words = len(words)         # ƒê·∫øm s·ªë t·ª´
    print("‚úÖ File '" + filename + "' c√≥ " + str(num_words) + " t·ª´.")
```

```python
def count_words(filename):
    """H√†m ƒë·∫øm s·ªë t·ª´ trong file vƒÉn b·∫£n"""
    try:
        with open(filename, 'r', encoding='utf-8') as f_obj:
            contents = f_obj.read()
    except FileNotFoundError:
        msg = "‚ùå File '" + filename + "' kh√¥ng t·ªìn t·∫°i."
        print(msg)
    else:
        words = contents.split()
        num_words = len(words)
        print("üìò File '" + filename + "' c√≥ " + str(num_words) + " t·ª´.")


# Danh s√°ch c√°c file c·∫ßn ƒë·∫øm
filenames = ['f1.txt', 'f2.txt', 'f3.txt']

# G·ªçi h√†m ƒë·ªÉ ƒë·∫øm t·ª´ trong t·ª´ng file
for filename in filenames:
    count_words(filename)
```

```python
import zipfile
import os

def compress(file_name, zip_name):
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        zipf.write(file_name, os.path.basename(file_name))
        print(f"‚úÖ ƒê√£ n√©n '{file_name}' th√†nh '{zip_name}'")

def decompress(zip_name, extract_to='.'):
    with zipfile.ZipFile(zip_name, 'r') as zipf:
        zipf.extractall(extract_to)
        print(f"‚úÖ ƒê√£ gi·∫£i n√©n '{zip_name}' v√†o '{extract_to}'")

# Ch∆∞∆°ng tr√¨nh ch√≠nh
if __name__ == "__main__":
    compress('example.txt', 'example.zip')
    decompress('example.zip', 'output_folder')
```
