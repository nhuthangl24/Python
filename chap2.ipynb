```python
def tao_TD(Max):
    d = {}
    for i in range(1, Max + 1):
        d[i] = i ** 2
    return d
def Print_Item(TD):
    for k, v in TD.items():
        print(k, v)
def Print_key(TD):
    for k in TD.keys():
        print(k)
def Print_value(TD):
    for v in TD.values():
        print(v)
max_val = int(input("Nhập chỉ số Max: "))
TD = tao_TD(max_val)
print("Các phần tử của từ điển là:")
Print_Item(TD)
print("Khóa của các phần tử từ điển:")
Print_key(TD)
print("Giá trị của các phần tử từ điển:")
Print_value(TD)
```
